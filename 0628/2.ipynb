```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
tf.__version__
```

```python
from tensorflow import keras

# 설정

model = keras.models.Sequential() # Sequential 이라는 모델을 쓰겠다.
model.add(keras.layers.Dense(1, input_shape=(3,))) # Dense라는 층을 만든다.
model.summary()
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
X = iris.data[:, :3]
y = iris.data[:, 3]

X.shape, y.shape
```

```python
model.compile(loss='mse', optimizer='sgd')
```

```python
model.fit(X, y, epochs=20, validation_split=0.25) 
# validation_split --> 25%를 검증용으로 빼놓겠다. 그러면 140중 115개로 러닝하고 25개로 검증한다.
# 발표를 하면 검증로스로 발표하는게 좋다. ( 검증이 중요하기 때문이다. )
```

```python
h = model.history.history
```

```python
plt.plot(h['loss'])
plt.plot(h['val_loss'])
plt.legend(['loss', 'val_loss'])
```

```python
model.fit(X, y, epochs=200, validation_split=0.25)
```

```python
h = model.history.history
plt.plot(h['loss'])
plt.plot(h['val_loss'])
plt.legend(['loss', 'val_loss'])
plt.ylim(0.02, 0.15)
plt.xlim(0, 200)
```
